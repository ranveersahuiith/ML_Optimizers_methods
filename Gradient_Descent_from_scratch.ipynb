```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.linspace(0,10,50) # random data of feature 1 data from 1-10 and total 50 datas

true_m, true_c = 2,1 # we know the actual line parameters slop = 2 and intercept = 1


Y = true_m*X + true_c + np.random.normal(0,2,X.shape) # actaul line will be with some noise

```

```python
plt.figure(figsize=(10,6)) # grid box in 2-d
plt.scatter(X,Y,label = "Data Points", color="green") # scatter plot to show datas
plt.title("How Data looks like") # title
plt.xlabel("X") # x_axis mark
plt.ylabel("Y") # y_axis mark
plt.legend() # marking the other stuffs
plt.show()
```

```python
m,c = 0.0,0.0 # actual GD start from here, assume m and c randomlly

learning_rate = 0.01 # set a Learning rate

iterations = 2000 # set interations
```

```python
loss_history  = [] # storing the loss history, it's on you!
```

```python
# actual GD begin
for i in range(iterations):
  Y_pred = m*X + c # for m and c, find Y_pred

  m_grad = (-2/len(X))*np.sum(X*(Y-Y_pred)) # grad_of m from loss_function
  c_grad = (-2/len(X))*np.sum(Y-Y_pred) # grad_of_c from loss function

  m = m - learning_rate*m_grad # m update
  c = c- learning_rate*c_grad # c update

  loss = np.mean((Y-Y_pred)**2) # again find loss for this iteration for different value of m and c.
  loss_history.append(loss) # store in loss_history

  if i%100==0:
    print(f"Iteration {i}: Loss = {loss:.4f}, m = {m:.4f}, c = {m:.4f}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,label = "Data Points", color="green")
plt.plot(X,m*X+c, label=f"Fitted line: y = {m:.2f}x + {c:.2f}", color = "red") # plot used for ploting something
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
